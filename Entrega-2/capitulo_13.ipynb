## 13.1 Aplicações de Sistemas de Controle com Computadores Digitais

Sistemas de controle que utilizam computadores digitais processam informações de sensores e decidem ações sobre atuadores para regular o comportamento de máquinas, processos ou dispositivos. Eles são encontrados em muitas áreas, como indústrias, usinagem, plantas químicas, controle de tráfego e veículos modernos, incluindo aviões e carros autônomos. A principal vantagem desses sistemas é a capacidade de processar sinais de entrada, executar algoritmos complexos e ajustar a saída de forma rápida e precisa.

Um computador digital típico de controle possui três elementos principais:
- Unidade Central de Processamento (CPU), que realiza cálculos e executa o programa de controle.
- Memória, que armazena o programa e os dados de sensores e atuadores.
- Interfaces de entrada e saída (I/O), que conectam o computador aos sensores que medem variáveis físicas e aos atuadores que aplicam correções no sistema.

Microcomputadores baseados em microprocessadores são frequentemente usados em sistemas de controle devido ao seu tamanho reduzido, baixo custo e grande capacidade de processamento. A escolha do computador depende da complexidade do processo a ser controlado, da quantidade de informações que devem ser armazenadas e da velocidade com que a resposta do sistema é necessária.

Um exemplo prático de aplicação de controle digital está nos aviões Boeing 757 e 767. Nesses aviões, sistemas eletrônicos monitoram continuamente os motores e outros parâmetros de voo. O computador ajusta automaticamente o empuxo dos motores e a trajetória do voo, garantindo eficiência no consumo de combustível e mantendo a segurança durante decolagem, voo e pouso. Essas funções integram sensores, atuadores e algoritmos digitais para realizar correções automáticas em tempo real.

Entre as vantagens de sistemas de controle digital estão:
1. Sensibilidade elevada nas medições, pois sensores digitais conseguem detectar pequenas variações nos sinais.
2. Uso de sinais digitais codificados, que são menos afetados por ruídos e interferências.
3. Capacidade de processar e integrar dados de múltiplos sensores e transdutores digitais.
4. Robustez contra ruído elétrico, resultando em respostas mais precisas.
5. Flexibilidade para modificar ou atualizar o algoritmo de controle apenas via software, sem necessidade de alterações físicas no hardware.

O controle digital permite que sistemas complexos operem de maneira confiável, eficiente e adaptável, sendo essencial em processos onde precisão e segurança são críticas.


## 13.2 Sistemas de Dados Amostrados

Um sistema de dados amostrados é um tipo de sistema em que as informações de entrada e saída são representadas de forma discreta, mesmo que o processo físico seja contínuo. Esses sistemas são a base do controle digital, pois permitem que computadores realizem cálculos e executem algoritmos de controle a partir de sinais que foram convertidos em valores discretos ao longo do tempo.

A operação desses sistemas envolve a coleta de amostras do sinal contínuo em intervalos regulares de tempo, chamados de período de amostragem, denotado por $$T$$. Assim, um sinal de referência contínuo $$r(t)$$ é transformado em uma sequência de valores $$r(kT)$$, onde $$k$$ é um número inteiro que indica a posição da amostra na sequência temporal. Esse processo de amostragem permite que o computador trate o sinal como uma série de pontos discretos, facilitando o processamento digital.

Um sistema de controle digital típico baseado em dados amostrados inclui três elementos principais:

1. Conversor analógico-digital (A/D): recebe o sinal contínuo do mundo físico e o converte em valores discretos que podem ser processados pelo computador. Esse passo é essencial para que os algoritmos digitais possam interpretar o comportamento do sistema.  

2. Computador digital: executa os cálculos de controle utilizando os valores discretos fornecidos pelo conversor A/D. O computador implementa algoritmos que determinam como o sistema deve reagir a cada instante, gerando sinais de controle apropriados.  

3. Conversor digital-analógico (D/A): transforma os sinais de controle calculados pelo computador, que estão em formato discreto, de volta para um sinal contínuo. Esse sinal contínuo é então aplicado aos atuadores do sistema, permitindo que o processo físico seja ajustado conforme necessário.  

O ciclo completo de um sistema de dados amostrados segue o seguinte fluxo: o sinal contínuo do ambiente é amostrado, processado pelo computador e reconvertido para atuar sobre o sistema físico. Esse método permite a implementação de estratégias de controle complexas, com flexibilidade e precisão, mantendo o desempenho mesmo em processos dinâmicos e variáveis.


## 13.3 A Transformada Z

A transformada Z é uma ferramenta matemática fundamental para estudar sistemas de controle digital que trabalham com sinais amostrados. Como computadores digitais processam dados apenas em instantes discretos de tempo, precisamos de um método para representar e analisar como esses sistemas respondem a sinais de entrada e como se comportam ao longo do tempo.

Se considerarmos um sinal contínuo $$r(t)$$ que é amostrado a intervalos regulares de tempo $$T$$, podemos escrever o sinal amostrado como $$r^*(t)$$. A transformada de Laplace deste sinal amostrado é expressa por:

$$
\mathcal{L}\{r^*(t)\} = \sum_{k=0}^{\infty} r(kT) e^{-ksT}
$$

A transformada Z surge a partir da relação entre o plano s da transformada de Laplace e o plano z, definida por:

$$
z = e^{sT}
$$

Essa relação permite mapear funções do domínio contínuo (s) para o domínio discreto (z). Assim, a transformada Z de uma função discreta $$f(kT)$$ é dada por:

$$
\mathcal{Z}\{f(t)\} = F(z) = \sum_{k=0}^{\infty} f(kT) z^{-k}
$$

A transformada Z possui propriedades semelhantes às da transformada de Laplace, incluindo linearidade, deslocamento e multiplicação por exponenciais. Ela é utilizada para analisar a estabilidade de sistemas discretos, determinar a resposta transitória e projetar controladores digitais. Tabelas de transformadas Z e suas propriedades auxiliam na aplicação prática desses cálculos.

### Transformada Z de Funções Exponenciais e Senoidais

Para ilustrar, podemos calcular a transformada Z de uma função exponencial decrescente:

$$
f(t) = e^{-at}
$$

Se amostrarmos essa função a intervalos $$T$$, obtemos a sequência:

$$
f(kT) = e^{-akT}
$$

Aplicando a definição da transformada Z, temos:

$$
F(z) = \sum_{k=0}^{\infty} f(kT) z^{-k} = \sum_{k=0}^{\infty} e^{-akT} z^{-k}
$$

Reconhecendo esta soma como uma série geométrica, podemos simplificar para:

$$
F(z) = \frac{1}{1 - e^{-aT} z^{-1}}, \quad |z| > e^{-aT}
$$

De forma similar, a transformada Z de funções senoidais amostradas pode ser obtida aplicando a definição, permitindo estudar respostas oscilatórias de sistemas discretos e analisar seu comportamento em frequência.

## 13.4 Análise de estabilidade no plano Z

Em sistemas de controle contínuos e lineares, a estabilidade é garantida quando todos os polos da função de transferência de malha fechada, $$T(s)$$, estão localizados no semiplano esquerdo do plano $$s$$. Ou seja, para que o sistema não oscile de forma indefinida nem cresça com o tempo, todos os polos devem ter parte real negativa ($$\sigma < 0$$).

Quando se trabalha com sistemas discretos, amostrados em intervalos de tempo $$T$$, o plano $$s$$ é mapeado para o plano $$z$$ usando a relação:

$$
z = e^{sT} = e^{(\sigma + j\omega)T}
$$

Essa relação implica que:

$$
|z| = e^{\sigma T} \quad \text{e} \quad \angle z = \omega T
$$

Portanto, os polos localizados no semiplano esquerdo do plano $$s$$ (onde $$\sigma < 0$$) correspondem a pontos dentro do círculo unitário no plano $$z$$, ou seja, onde a magnitude de $$z$$ satisfaz:

$$
0 < |z| < 1
$$

De forma análoga, os polos no eixo imaginário do plano $$s$$ (onde $$\sigma = 0$$) correspondem a pontos sobre o círculo unitário no plano $$z$$.

A regra geral de estabilidade para sistemas discretos é: **um sistema amostrado é estável se todos os polos da função de transferência de malha fechada, $$T(z)$$, estiverem localizados dentro do círculo unitário do plano $$z$$**.

Para verificar a estabilidade, analisamos a equação característica do sistema:

$$
q(z) = 1 + G(z) = 0
$$

As raízes dessa equação, chamadas polos do sistema, determinam seu comportamento dinâmico. Por exemplo, para um sistema de segunda ordem com coeficientes reais, os critérios necessários e suficientes para que todas as raízes estejam dentro do círculo unitário são:

$$
|q(0)| < 1, \quad q(1) > 0, \quad q(-1) > 0
$$

Essas condições permitem confirmar se o sistema discreto permanecerá estável, sem oscilações crescentes ou divergência ao longo do tempo.


## 13.5 Desempenho de um sistema de segunda ordem com dados amostrados

Para analisar o desempenho de um sistema de segunda ordem amostrado, consideremos um processo típico a ser controlado descrito pela função de transferência contínua:

$$
G_p(s) = \frac{K}{s(\tau s + 1)}
$$

onde $$K$$ é o ganho do sistema e $$\tau$$ é a constante de tempo do processo.

Quando utilizamos um extrapolador de ordem zero para amostrar o sistema, obtemos a função de transferência discreta $$G(z)$$ em termos do período de amostragem $$T$$:

$$
G(z) = \frac{K \left[(z - E)\left(T - \tau (z - 1)\right) + \tau (z - 1)^2\right]}{(z - 1)(z - E)}
$$

com

$$
E = e^{-T / \tau}
$$

O desempenho e a estabilidade do sistema discreto podem ser estudados analisando a **equação característica**, que define os polos do sistema de malha fechada:

$$
q(z) = z^2 + z \left\{ K \left[T - \tau (1 - E)\right] - (1 + E) \right\} + K \left[T (1 - E) - \tau E \right] + E = 0
$$

Como este é um polinômio quadrático com coeficientes reais, as **condições necessárias e suficientes** para que todas as raízes fiquem dentro do círculo unitário no plano $$z$$, garantindo estabilidade do sistema discreto, são:

$$
|q(0)| < 1, \quad q(1) > 0, \quad q(-1) > 0
$$

Estas condições permitem avaliar se o sistema controlado permanecerá estável ao longo do tempo, mesmo após a discretização do processo e o uso do extrapolador de ordem zero. O comportamento dinâmico, incluindo resposta transitória e velocidade de convergência, depende tanto do ganho $$K$$ quanto do período de amostragem $$T$$.


## 13.6 Sistemas a malha fechada com compensação digital por computador

Um sistema de controle amostrado em malha fechada pode utilizar um computador digital para melhorar seu desempenho, ajustando a resposta do sistema de forma mais precisa. A função de transferência do sistema compensado em malha fechada é dada por:

$$
\frac{Y(z)}{R(z)} = T(z) = \frac{G(z) D(z)}{1 + G(z) D(z)}
$$

onde $$G(z)$$ representa a função de transferência do processo discretizado e $$D(z)$$ é a função de transferência do compensador digital implementado pelo computador. O computador calcula a ação de controle a partir do erro entre a referência $$R(z)$$ e a saída medida $$Y(z)$$, ou seja:

$$
\frac{U(z)}{E(z)} = D(z)
$$

Para ilustrar, considere um exemplo prático de projeto de sistema discreto com compensador digital. Suponha que o processo a ser controlado seja de segunda ordem, com retenção de ordem zero e período de amostragem $$T = 1$$, descrito por:

$$
G_p(s) = \frac{1}{s(s + 1)}
$$

Para melhorar o desempenho do sistema, seleciona-se um compensador digital com função de transferência:

$$
D(z) = \frac{1.359 (z - 0.3678)}{z + 0.240}
$$

Com isso, a função de transferência do sistema compensado em malha fechada torna-se:

$$
\frac{Y(z)}{R(z)} = \frac{K G(z) D(z)}{1 + K G(z) D(z)}
$$

Neste caso, o ganho $$K$$ do sistema ajusta a magnitude da resposta, enquanto o compensador digital $$D(z)$$ atua para melhorar a velocidade, reduzir o overshoot e garantir estabilidade no plano $$z$$. Este tipo de projeto permite que sistemas discretos alcancem desempenho próximo ao desejado, mesmo considerando a discretização e os atrasos introduzidos pela amostragem.


## 13.7 Implementação de Controladores Digitais

Um dos controladores mais utilizados em sistemas discretos é o controlador PID, cuja função de transferência no domínio contínuo é dada por:

$$
\frac{U(s)}{X(s)} = G_c(s) = K_1 + \frac{K_2}{s} + K_3 s
$$

onde $$K_1$$ é o ganho proporcional, $$K_2$$ é o ganho integral e $$K_3$$ é o ganho derivativo.

Para implementar esse controlador em um computador digital, precisamos aproximar as operações contínuas de derivada e integral por expressões discretas. Isso permite que o controlador seja calculado em cada instante de amostragem.

A derivada em relação ao tempo pode ser aproximada usando a **diferença atrasada**, que considera a variação do sinal entre duas amostras consecutivas:

$$
u(kT) = \frac{dx}{dt} \Big|_{t=kT} \approx \frac{1}{T} \left[ x(kT) - x((k-1)T) \right]
$$

Aplicando a transformada Z nesta aproximação, obtemos a forma discreta do termo derivativo:

$$
U(z) = \frac{1 - z^{-1}}{T} X(z) = \frac{z - 1}{T z} X(z)
$$

Dessa forma, o controlador derivativo pode ser implementado diretamente em um computador digital, utilizando apenas os valores discretos da entrada do sistema. A mesma abordagem pode ser estendida para os termos proporcional e integral, permitindo a implementação completa de um controlador PID digital.

```python
import numpy as np
import matplotlib.pyplot as plt
import control as ctl
from control import pade, minreal

num_cont = [1]
den_cont = [1, 1, 0]
Ts = 1

sys_cont = ctl.TransferFunction(num_cont, den_cont)
sys_disc = ctl.c2d(sys_cont, Ts, method="zoh")

print("Sistema discretizado com ZOH:")
print(sys_disc)

num_disc = [0.3678 + 0.2644]
den_disc = [1 - 1 + 0.6322]
sys_disc_example = ctl.TransferFunction(num_disc, den_disc)

tempo_sim = np.arange(0, 8, 0.1)
saida_deg, tempo_saida = ctl.step_response(sys_disc_example, tempo_sim)

plt.figure(figsize=(7,4))
plt.plot(tempo_saida, saida_deg, 'o-', color='blue', markersize=5, linewidth=1.5)
plt.title('Resposta ao Degrau - Sistema Discreto (pontos conectados)')
plt.xlabel('Tempo [s]')
plt.ylabel('Amplitude')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

num_sys = np.array([1])
den_sys = np.array([1, 1, 0])

num_pade, den_pade = pade(1, 2)
num_pade_disc = np.array(den_pade) - np.array(num_pade)
den_pade_disc = np.convolve(np.array([1, 0]), den_pade)

sys_pade = minreal(ctl.TransferFunction(num_pade_disc, den_pade_disc))
sys_serie = ctl.series(sys_pade, ctl.TransferFunction(num_sys, den_sys))
sys_feedback = ctl.feedback(sys_serie)

saida_fb, tempo_fb = ctl.step_response(sys_feedback)

plt.figure(figsize=(7,4))
plt.plot(tempo_fb, saida_fb, '--', color='red', linewidth=2)
plt.title('Resposta ao Degrau - Pade com Feedback (linha tracejada)')
plt.xlabel('Tempo')
plt.ylabel('y(t)')
plt.grid(True, linestyle=':')
plt.axis([0, 20, 0, 1.5])
plt.show()

num_rlocus = np.array([0.3678, 0.2644])
den_rlocus = np.array([1.0, -0.76, -0.24])
sys_rlocus = ctl.TransferFunction(num_rlocus, den_rlocus)

plt.figure(figsize=(6,6))
rlist, klist = ctl.rlocus(sys_rlocus, xlim=[-2, 2], ylim=[-2, 2], plot=False)
plt.plot(np.real(rlist), np.imag(rlist), 'b.', markersize=6)
plt.title('Lugar das Raízes com destaque')
plt.xlabel('Parte Real')
plt.ylabel('Parte Imaginária')

theta = np.linspace(0, 2*np.pi, 100)
plt.plot(np.cos(theta), np.sin(theta), '--', color='green')
plt.grid(True)
plt.axis('equal')
plt.show()

x_vals = np.arange(-1, 1.01, 0.01)
y_vals = np.sqrt(1 - x_vals**2)

plt.figure(figsize=(6,6))
plt.fill_between(x_vals, y_vals, -y_vals, color='lightblue', alpha=0.3)
plt.plot(x_vals, y_vals, 'b', label='Limite superior')
plt.plot(x_vals, -y_vals, 'b', label='Limite inferior')
plt.title('Círculo Unitário - Região de Estabilidade')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.axis('equal')
plt.legend()
plt.show()
```
